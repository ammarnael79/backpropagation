```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

inputs = np.array([[0.05, 0.10]])
expected_output = np.array([[0.01, 0.99]])

weights_input_hidden = np.array([[0.15, 0.25], [0.20, 0.30]])
weights_hidden_output = np.array([[0.40, 0.50], [0.45, 0.55]])

bias_hidden = np.array([0.35, 0.35])
bias_output = np.array([0.60, 0.60])

lr = 0.001

hidden_input = np.dot(inputs, weights_input_hidden) + bias_hidden
hidden_output = sigmoid(hidden_input)

final_input = np.dot(hidden_output, weights_hidden_output) + bias_output
final_output = sigmoid(final_input)

error = 0.5 * (expected_output - final_output) ** 2

output_error = expected_output - final_output
output_delta = output_error * sigmoid_derivative(final_output)

hidden_error = output_delta.dot(weights_hidden_output.T)
hidden_delta = hidden_error * sigmoid_derivative(hidden_output)

weights_hidden_output += hidden_output.T.dot(output_delta) * lr
weights_input_hidden += inputs.T.dot(hidden_delta) * lr

bias_output += np.sum(output_delta, axis=0) * lr
bias_hidden += np.sum(hidden_delta, axis=0) * lr

print("Updated weights (input to hidden):\n", weights_input_hidden)
print("Updated weights (hidden to output):\n", weights_hidden_output)
print("Updated bias (hidden):\n", bias_hidden)
print("Updated bias (output):\n", bias_output)



```
